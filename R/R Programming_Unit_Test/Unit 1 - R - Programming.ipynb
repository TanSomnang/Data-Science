## 2. Solve Linear Equation using Gaussian Elimination

```r
library(matlib)
Ax <- matrix(c(1, 2, 3, 
                        2, 1, 2, 
                        3, 3, 1), 3, 3, byrow=TRUE)
y <- c(9, -3, 5)
result <- gaussianElimination(Ax, y)
paste('x = ',result[10],' y = ', result[11], ' z = ',result[12])
```

### b. Explain the Unit and Necessity of each variable

- This dataset contains the information of covid-19 cases which happen globally by December 4<sup>th</sup> 2020
- Source: World Health Organization
- Dataset size: 236 Rows ,11 Columns
- A row of this dataset indicates total of covid-19 cases of a specific country and globally
- Columns description and data type:
    1. Column 1: Name (string)
    2. Column 2: WHO Region (string)
    3. Column 3: TotalCases (double)
    4. Column 4: Cases - cumulative total per 1 million population (double)
    5. Column 5: Cases - newly reported in last 7 days (double)
    6. Column 6: Cases - newly reported in last 24 hours (double)
    7. Column 7: TotalDeaths (double)
    8. Column 8: Deaths - cumulative total per 1 million population (double)
    9. Column 9: Deaths - newly reported in last 7 days (double)
    10. Column 10: Deaths - newly reported in last 24 hours (double)
    11. Column 11: Transmission Classification (string)

```r
library(tidyverse)
dataset <- read_csv('WHO COVID-19.csv')
print(paste('Total Rows: ', dim(dataset)[1]))
print(paste('Total Columns: ', dim(dataset)[2]))
```

### c. Find missing values and replace with mean (Tidy Dataset)

#### - Find missing values

```r
missing = colnames(dataset)[ apply(dataset, 2, anyNA) ]
print( missing)
```

#### - Replace with mean (Tidy Dataset)

```r
mean_casesPerMillion =  apply(dataset[,'Cases - cumulative total per 1 million population'], 2, mean, na.rm =  TRUE)
mean_deathPerMillion = apply(dataset[,'Deaths - cumulative total per 1 million population'], 2, mean, na.rm =  TRUE)
mean_casesPerMillion
mean_deathPerMillion
dim(dataset)
```

- String cannot be replaced with mean value,

```r
drop_empty <- dataset % > % na.omit()
dim(drop_empty)
```

### d. Generate the 2 new variables (Mean, Median)

- Var1 = mean_casesPerMillion (Mean)
- Var2 = Cases - cumulative total (Median)

```r
new_variables <- mutate(drop_empty, Var1 <- mean_casesPerMillion, Var2 <- apply(dataset[,'Cases'], 2, median, na.rm=TRUE))
new_variables
```

### e. Rename 2 existing variables

```r
rename <- rename(new_variables,
    Var1 = Mean ,
    Var2 = Median
)
new_variables
```

### f. Create a plot using following instruction:

#### i. Choose x and y axis
#### ii. geom_point() - specify the parameters, size : 5, color: red, alpha: 1⁄5
#### iii. Use Facet grid, cartesian coordinates & geom_smooth()
#### iv. Assign the title to x, y and graph
#### v. Export the graph to your working directory with the title called “covid_19_dataset.png”


```r
europe = dataset[dataset$'WHO Region' == "Europe",]
ggplot(europe, aes(Deaths, Cases)) +
    geom_point(size = 5, color = 'red', alpha = 0.5) + 
    geom_smooth() + ggtitle("Cases VS Deaths in Europe") + xlab("Deaths") + ylab("Cases") + 
    ggsave(filename = "covid_19_dataset.png", units = "mm", width = 150, height = 100)
```
