## 11. Calculate BMI

a). If your BMI is less than 18.5, it falls within the underweight range.<br>
b). If your BMI is 18.5 to <25, it falls within the normal.<br>
c). If your BMI is 25.0 to <30, it falls within the overweight range.<br>
d). If your BMI is 30.0 or higher, it falls within the obese range.

```r
BMI.function <- function(weight, height){
    bmi <- weight / ((height / 100) ^ 2)
    result <- as.integer(bmi)
    cat("BMI = ", result)
    if(result < 18.5){
        "Underweight"
    }else if(result >= 18.5 && result < 25){
        result
        "Normal"
    }else if(result >= 25 && result < 30){
        result
        "Overweight"
    }else{
        result
        "Obese"
    }
}
weight <- readline(prompt = "Enter your weight in kg: ")
weight <- as.integer(weight)
height <- readline(prompt = "Enter your height in cm: ")
height <- as.integer(height)
BMI.function(weight, height)
```

## 12. Difference Between List and Data frame

- List is a vector with possible heterogeneous elements <br>
- The elements of a list can be numeric vectors, character vectors, matrices, arrays, and lists<br>
- 1 dimension<br>
- Create by using list() function<br>

```r
myList <- list(a = 1:10, b = "def", c(TRUE, FALSE, TRUE))
myList
```

### Data Frame

-  Data Frame is a list with possible heterogeneous vector elements of the same length<br>

## 13. Difference Between repeat and while loop

### Repeat

- Repeat: it is a simple loop that will run the same statement or a group of statements repeatedly until the stop condition has been encountered<br>
- Repeat loop does not have any condition to terminate the loop, a programmer must specifically place a condition within the loop’s body and use the declaration of a break statement to terminate this loop<br>
- If no condition is present in the body of the repeat loop then it will iterate infinitely

```r
val = 1
repeat 
{
    print(val) 
    val = val + 1
    if(val > 5)  
    {
        break
    } 
} 
```

### While

- It is a type of control statement which will run a statement or a set of statements repeatedly unless the given condition becomes false<br>
- It is also an entry controlled loop, in this loop the test condition is tested first, then the body of the loop is executed, the loop body would not be executed if the test condition is false

```r
val = 1
while (val <= 5)  
{
    print(val)
    val = val + 1
} 
```
